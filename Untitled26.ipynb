```python
import matplotlib.pyplot as plt

# Data
years = list(range(2000, 2019))
numbers = [
    187000, 178000, 155000, 141000, 127000,
    117000, 125000, 113000, 94000, 96000,
    102000, 107000, 112000, 112000, 108000,
    107000, 110000, 112000, 112000
]

# Create the bar chart
plt.figure(figsize=(12, 8))
plt.bar(years, numbers, color='teal')
plt.title('Naturalized Refugees from 2000-2018')
plt.xlabel('Year')
plt.ylabel('Number of Refugees')
plt.xticks(years, rotation=45)
plt.grid(axis='y')

# Show the chart
plt.tight_layout()
plt.show()
```
